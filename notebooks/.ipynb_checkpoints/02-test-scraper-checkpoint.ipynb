```python
import pandas as pd
import requests
import time

from datetime import datetime
```

```python
def call_bike_api():
    
    url = 'https://gbfs.capitalbikeshare.com/gbfs/en/free_bike_status.json'
    
    r = requests.get(url)
    j = r.json()
    b = j['data']['bikes']
    
    bikes = pd.json_normalize(b)
    bikes['timestamp'] = datetime.now().strftime("%H:%M:%S")
    
    return bikes   
```

```python
def call_station_status_api():
    
    url = 'https://gbfs.capitalbikeshare.com/gbfs/en/station_status.json'
    
    r = requests.get(url)
    j = r.json()
    b = j['data']['stations']
    
    bikes = pd.json_normalize(b)
    bikes['timestamp'] = datetime.now().strftime("%H:%M:%S")
    
    return stations 
```

```python
def scrape_bikes(interval):
    
    bike_data = pd.DataFrame()
    station_data = pd.DataFrame()
    timer = 0
    
    while timer < 30:
        
        current_bikes = call_bike_api()
        current_stations = call_station_api()
        
        bike_data = pd.concat((bike_data, current_bikes))
        station_data = pd.concat((station_data, current_stations))
        
        bike_data.to_csv('../raw_data/bike_data.csv')
        station_data.to_csv('../raw_data/station_data.csv')
        
        timer = timer + interval
        time.sleep(interval)
```
